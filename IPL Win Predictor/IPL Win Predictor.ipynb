```python
new_df = second_inn[[ "BattingTeam", "City","Bowling_team", "total_run", "runs_left", "balls_remaining", "current_wicket", "current_run_rate", "required_run_rate","Match_result"]]
```

```python
final_df = new_df.sample(new_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.rename(columns = {"BattingTeam" : "batting_team", "City" : "city", "Bowling_team" : "bowling_team", "total_run" : "total_run_x", "balls_remaining" : "balls_left", "current_wicket" : "wicket", "current_run_rate" : "crr","required_run_rate":"rrr", "Match_result":"result"}, inplace = True)
```

```python
final_df.columns
```

```python
final_df.dropna(inplace = True)
```

```python
final_df = final_df[final_df["balls_left"] != 0]
```

```python
from sklearn.model_selection import train_test_split

X = final_df.iloc[:, :-1]
y = final_df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
X_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([('trf', OneHotEncoder(sparse = False, drop = "first"), ["batting_team", "city", "bowling_team"]) ], remainder = "passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
final_df.isnull().sum()
```

```python
final_df.isnull().sum()
```

```python
pipe = Pipeline(steps = [
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)      # Implementing Logistic Regeression
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
pipe2 = Pipeline(steps = [
    ('step1', trf),
    ('step2', RandomForestClassifier())
])
```

```python
pipe2.fit(X_train, y_train)        # Implementing Random Forest Model
```

```python
y_predicted = pipe2.predict(X_test)
```

```python
accuracy_score(y_test, y_predicted)
```
